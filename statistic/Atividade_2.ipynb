```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, laplace, norm
```

```python
s= 1000
laplace_samples = np.random.laplace(loc=3.72, scale=0.17, size=s)
chisquare_samples = np.random.chisquare(df=2, size=1000)

fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)
laplace_bins = int(2 * len(laplace_samples)**(1/3))
axes[0].hist(laplace_samples, bins=laplace_bins, density=True, alpha=0.6, color='blue')

laplace_mean, laplace_scale = np.mean(laplace_samples), np.std(laplace_samples)
x_laplace = np.linspace(min(laplace_samples), max(laplace_samples), s)

laplace_pdf = laplace.pdf(x_laplace, loc=laplace_mean, scale=laplace_scale)
axes[0].plot(x_laplace, laplace_pdf, 'r-', lw=2)

axes[0].set_title('Distribuição Laplace')
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Densidade')

chisquare_bins = int(2 * len(chisquare_samples)**(1/3))
axes[1].hist(chisquare_samples, bins=chisquare_bins, density=True, alpha=0.6, color='green')

df_chisquare = 2  # graus de liberdade para chi quadrado
x_chisquare = np.linspace(min(chisquare_samples), max(chisquare_samples), s)

chisquare_pdf = chi2.pdf(x_chisquare, df_chisquare)
axes[1].plot(x_chisquare, chisquare_pdf, 'r-', lw=2)

axes[1].set_title('Distribuição Qui-quadrado')
axes[1].set_xlabel('Valor')

plt.tight_layout()
plt.show()
```

```python
np.mean(chisquare_samples)
```

```python
np.mean(laplace_samples)
```

## Demonstrando TLC

### Parâmetros da distribuição Laplaciana
* Loc (média) = 3.72 <br>
* Escala (dispersão) = 0.17

### Parâmetros da distribuição Chi Quadrado
* Número de graus de liberdade = 0.55

```python
def plot_samples(data):
  sample_sizes = [5, 30, 100, 500, 1000]
  n_samples = 1000

  fig, axes = plt.subplots(1, 5, figsize=(20, 6), sharey=True)
  rice_bins = int(2 * len(data)**(1/3))

  for i, size in enumerate(sample_sizes):
      sample_means = [np.mean(np.random.choice(data, size=size)) for _ in range(n_samples)]

      axes[i].hist(sample_means, bins=rice_bins, density=True, alpha=0.6, color='blue')

      mu, std = np.mean(sample_means), np.std(sample_means)
      x = np.linspace(min(sample_means), max(sample_means), n_samples)
      axes[i].plot(x, norm.pdf(x, mu, std), 'r-', lw=2)

      axes[i].set_title(f'Sample Size = {size}')
      axes[i].set_xlabel('Mean')
      axes[i].set_ylabel('Density')

  plt.tight_layout()
  plt.show()
```

```python
# Parâmetros da distribuição Laplace
loc = 3.72
scale = 0.17

# Parâmetro da distribuição chisquare
df=0.55

laplace_data = np.random.laplace(loc=loc, scale=scale, size=10000)
chisquare_data = np.random.chisquare(df=df, size=10000)
```

```python
plot_samples(laplace_data)
```

```python
np.mean(laplace_data)
```

```python
plot_samples(chisquare_data)
```

```python
np.mean(chisquare_data)
```
